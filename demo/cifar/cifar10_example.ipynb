import packages

```python
import cifar10vgg
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import cv2
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
%matplotlib inline  
```

load pretrained model 

```python
model = cifar10vgg.cifar10vgg(train=False)
```

def class array

```python
str_class=[]
str_class.append("airplane")
str_class.append("automobile")
str_class.append("bird")
str_class.append("cat")
str_class.append("deer")
str_class.append("dog")
str_class.append("frog")
str_class.append("horse")
str_class.append("ship")
str_class.append("truck")
```

set plot scale

```python
import matplotlib as mpl
default_figsize = mpl.rcParamsDefault['figure.figsize']
times = 0.2
mpl.rcParams['figure.figsize'] = [val * times for val in default_figsize]
```

predict img class and show image

```python
def predict_img(img_path):
    img = cv2.imread(img_path)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)
    
#     print("this img {} may be [{}] ".format(img_path, str_class[int(model.predict_class(x))]))
    plt.title("This image {} may be [{}] ".format(img_path, str_class[int(model.predict_class(x))]))
    plt.imshow(img)
```
